## Lists

- A list is an object that holds a collection of objects.
- Represents a sequence of data.
- Can contain every data type, but should be homogeneous.

```python
# Creating a list, 2 ways
list1 = ["H", "E", "L", "L", "O"]
list2 = list("HELLO")
list_number = [1, 2, 3]

# List slicing
print(list1[:2]) # index 0 to 2 --> index 2 not included prints ["H", "E"]
print(list2[-3:-2]) # index -3 to -2 --> index 2 not included prints ["L"]

# List funcs and methods

length1 = len(list1) # returns number of items -> 5
sum1 = sum(list_number) # returns sum of items -> 6
min1 = min(list_number) # returns min -> 1
max1 = max(list_number) # returns max -> 3

list1.append("!") # adds "!" at the end of the list
list1.sort()
list1.count("L") # Counts the occurences of "L" -> 2
list1.index("H" ) # index "H" -> 0
list1.reverse()
list1.remove("H") # remove first occurence of "H"
list1.pop(1) # removes the index 1 returns its value
list1.insert(0, "H") # adds "H" to index 0
```

## Sets

- In mathematics, a set is a collection of objects, potentially of many different type
- No two identical elements.
- No order

```python
# Creating a set

set1 = set('abc')
set2 = {'a', 'b', 'c'}
set3 = {'a', 1, True}

# Set operators

a_set = set("abcd")
b_set = set("cdef")

intersection_a_b = a_set & b_set
intersection2_a_b = b_set.intersection(a_set) # both prints {'c','d'}

difference = a_set - b_set
difference2 = b_set.difference(a_set) # both prints {'e','f'}

union = a_set | b_set
union2 = b_set.union(a_set) # both {'a', 'b', 'c', 'd', 'e', 'f'}

symmetric_dif = a_set ^ b_set
symmetric_dif2 = b_set.symmetric_difference(a_set) # both {'a','b','e','f'}

# Subset "<=" and superset ">="

small_set = set("abc")
big_set = set("abcdef")

is_subset = small_set <= big_set # True
is_superset = big_set >= small_set # True
```

## Recursion in Python



### Key Concepts

- **Base Case**
  - condition that stops the recursion. Without it, the function would keep calling itself indefinitely and cause a stack overflow.
  - Every recursive func needs a base case
- **Recursive Case**
  - This is the part of the function where it calls itself
  - Every time the recursive case is called, it MUST have different input

```python
# Recursion Example in Python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:
        return n * factorial(n - 1)  # Recursive case

factorial(3) # Calling the recursive function the first time
```
